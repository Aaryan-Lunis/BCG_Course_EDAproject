
# Financial Data Analysis: Microsoft, Tesla, Apple (2021–2023)

This notebook analyzes key financial metrics for Microsoft, Tesla, and Apple over the last three fiscal years (2021–2023). The goal is to extract insights and trends to inform the development of an AI-powered financial chatbot.

## Key Metrics Analyzed:
- Total Revenue
- Net Income
- Total Assets
- Total Liabilities
- Cash Flow from Operating Activities


```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")
```

```python

# Load the CSV file
df = pd.read_csv('financial_data.csv')

# Display the first few rows
df.head()
```


## Data Overview

We begin by examining the raw financial data from 2021 to 2023.


```python

# Calculate year-over-year growth for each company
df.sort_values(by=['Company', 'Fiscal Year'], inplace=True)
df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income'].pct_change() * 100
df['Assets Growth (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby('Company')['Cash Flow from Operating Activities'].pct_change() * 100

df
```


## Trend Analysis

We visualize and analyze the financial trends across companies and years.


```python

# Plot Revenue Growth
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Fiscal Year', y='Total Revenue', hue='Company', marker='o')
plt.title('Total Revenue Over Time')
plt.ylabel('Revenue (in millions USD)')
plt.show()
```

```python

# Plot Net Income
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Fiscal Year', y='Net Income', hue='Company', marker='o')
plt.title('Net Income Over Time')
plt.ylabel('Net Income (in millions USD)')
plt.show()
```


## Summary of Findings

- **Microsoft** showed consistent growth in revenue and stable net income over the three years.
- **Tesla** had the most dramatic increase in net income and operating cash flow, reflecting its growth phase.
- **Apple** maintained strong revenue and profit but showed a slight dip in 2023 revenue.
- All companies increased their assets year-over-year, with liabilities growing at a slower rate, improving financial leverage.

These trends can help inform responses in an AI chatbot that answers questions on financial health and performance of major companies.
